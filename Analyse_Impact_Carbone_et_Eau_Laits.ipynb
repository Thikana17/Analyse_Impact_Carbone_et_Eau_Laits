```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Création des données (On va droit au but)
data = {
    'Type de Lait': ['Vache', 'Amande', 'Avoine', 'Soja'],
    'Emissions_CO2_kg': [3.2, 0.7, 0.9, 1.0],
    'Eau_litres': [628, 371, 48, 28]
}
df = pd.DataFrame(data)

# 2. Configuration du style graphique (Mode sombre/tech)
sns.set_theme(style="darkgrid")
plt.figure(figsize=(10, 6))

# 3. Création du nuage de points (Scatterplot)
sns.scatterplot(
    data=df,
    x='Emissions_CO2_kg',
    y='Eau_litres',
    hue='Type de Lait',
    s=800, # Bulles bien grosses
    palette=['#e74c3c', '#e67e22', '#2ecc71', '#3498db'] # Rouge, Orange, Vert, Bleu
)

# 4. Ajout du texte directement à côté de chaque bulle
for i in range(df.shape[0]):
    plt.text(df['Emissions_CO2_kg'][i] + 0.08, df['Eau_litres'][i] - 15, 
             df['Type de Lait'][i], 
             fontsize=12, weight='bold', color='black')

# 5. Titres et axes pour raconter l'histoire
plt.title("Le Piège du Lait d'Amande : L'impact caché sur l'Eau", fontsize=16, weight='bold', pad=20)
plt.xlabel("Émissions de CO2 (kg) ➡️", fontsize=12, weight='bold')
plt.ylabel("Consommation d'Eau Douce (Litres) ⬆️", fontsize=12, weight='bold')
plt.legend([],[], frameon=False) # On masque la légende par défaut

# 6. Sauvegarde en haute qualité (HD) et affichage
plt.savefig("dataviz_lait_amande.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# 1. Tes données complètes (CO2, Eau, et Terre)
data = {
    'Type de Lait': ['Vache', 'Amande', 'Avoine', 'Soja'],
    'Emissions_CO2_kg': [3.2, 0.7, 0.9, 1.0],
    'Eau_litres': [628, 371, 48, 28],
    'Terre_m2': [8.9, 0.5, 0.8, 0.7]
}
df_new = pd.DataFrame(data)

# 2. Création du graphique interactif Plotly
fig = px.scatter(
    df_new, 
    x='Emissions_CO2_kg', 
    y='Eau_litres', 
    size='Terre_m2',      # La taille des bulles = la déforestation/terre utilisée
    color='Type de Lait', # Une couleur automatique par lait
    title="Le Piège du Lait d'Amande : L'impact caché sur l'Eau",
    labels={
        'Emissions_CO2_kg': "Émissions CO2 (kg)", 
        'Eau_litres': "Consommation d'Eau (Litres)", 
        'Terre_m2': "Terre utilisée (m²)"
    },
    size_max=45           # Pour bien exagérer la taille de la bulle de la vache
)

# 3. Design : on met un fond blanc très propre pour LinkedIn
fig.update_layout(
    xaxis_title_font=dict(size=14, weight='bold'),
    yaxis_title_font=dict(size=14, weight='bold'),
    title_font=dict(size=18, weight='bold'),
    template="plotly_white" 
)

# On ajoute un petit contour noir autour des bulles pour faire "pro"
fig.update_traces(marker=dict(line=dict(width=2, color='DarkSlateGrey')))

# 4. Affichage de la magie !
fig.show()
```

```python
# Petit calcul
ratio_eau = df_new.loc[df_new['Type de Lait'] == 'Amande', 'Eau_litres'].values[0] / df_new.loc[df_new['Type de Lait'] == 'Avoine', 'Eau_litres'].values[0]

print(f"Bilan Data : Le lait d'amande consomme {ratio_eau:.1f} fois plus d'eau que le lait d'avoine !")
```

```python
# On utilise toujours Plotly (px) et notre tableau (df_new)
# On trie les données pour que le pire soit à gauche
df_trie = df_new.sort_values('Emissions_CO2_kg', ascending=False)

fig2 = px.bar(
    df_trie, 
    x='Type de Lait', 
    y='Emissions_CO2_kg',
    color='Type de Lait',
    title="Le Vrai Monstre du Climat : Les Émissions de CO2",
    text='Emissions_CO2_kg', # Affiche le chiffre exact au-dessus de la barre
    labels={'Emissions_CO2_kg': "Émissions de CO2 (kg)"}
)

# On met le texte en dehors de la barre pour que ce soit plus lisible
fig2.update_traces(textposition='outside', marker=dict(line=dict(width=1, color='black')))
fig2.update_layout(template="plotly_white")

fig2.show()
```

```python
fig3 = px.pie(
    df_new, 
    names='Type de Lait', 
    values='Terre_m2',
    title="Répartition de l'Espace Terrestre Nécessaire (Déforestation)",
    hole=0.4 # Ça transforme le camembert en "Donut", c'est beaucoup plus moderne et stylé !
)

fig3.show()
```
